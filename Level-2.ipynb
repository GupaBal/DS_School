# apply

```python
def is_vip(price):
    if price >= 500:
        return "VIP"
    else:
        return "Non-VIP"

    return price

order["price"].apply(is_vip)
```

```python
#컬럼이 하나가 아니라 여러가지를 사용하려 할떄 axis 옵션을 준다.

def is_vip(row):
    if row["state"] == "confirmed" and  row["price"] >=500:
        return "VIP"
    else:
        return "Non-VIP"

order.apply(is_vip, axis = "columns")
```

## 2-12 실제 많이 쓰는 다른데이터를 합쳐서 사용하기 pd.merge를 사용한다


```python
# 유저정보 가져오기
url = "https://bit.ly/dsa-02-user"

user = pd.read_csv(url)
user
```

```python
# 상품정보 가져오기 
url = "https://bit.ly/dsa-02-product"

product = pd.read_csv(url)
product
```

```python
# 유저정보와 상품정보를 합칠떄~!! 우선 공통되는 컬럼을 찾는다~!! 보통 KEY라고 부른다. 
#이 두개를 합치고 싶을때 pd.merge를 사용한다

pd.merge(order, user)   
```

```python
# order를 Left로  user를 Right로 간주한다. 
pd.merge(order, user, left_on = "user_id", right_on = "id")
```

### merge 옵션 왼쪽과 오른쪽을 합칠때  Key 가 다르면 Left On right On 으로 했는데..
### 가령 pd.merge(order, user, left_on = "user_id", right_on = "id") 이렇게 

```python
user["user_id"] = user.index
user
# user_id로 이름을 똑같이 맞춰주는 방법도 있다 
```

```python
pd.merge(order, user, on = "user_id")
```

### 2- 14 merge  order 결제정보 product 상품정보 

```python
pd.merge(order, product, left_on = "product_id", right_on ="id") # Key를 셋팅했을 경우 
# Key가 아니지만 똑같은 이름이 겹치면  price_x price_y 로 표시된다 이럴땐 하나만 살린다.  
```

```python
pd.merge(order, product) # Key를 셋팅하지 않을 경우 공통인 price로 합쳐버림 
```

### 2- 15 merge 마지막 Default
### how = "inner" >> 교집합  겹치는것만 남기고 나머지 버린다. 
### how = "outer"  >> 합집합 겹치지 않는것 모두 사용 
### how = "left"  >>  왼쪽은 살리고 오른쪽은 버린다. 
### how = "right"  >>  오른쪽을 살리고 왼쪽은 버린다. 
### 어떤 분석을 할때 분석의 메인이 되는 주제와 데이터를 생각하고 
### 이때, 메인데이터를 왼쪽에 놓고 무조건 left merge로 하는걸 추천 왠만하면 맞음 

```python
pd.merge(order,user,on = "user_id", how = "inner")
```

```python
pd.merge(order,user,on = "user_id", how = "outer")
```

```python
pd.merge(order,user,on = "user_id", how = "left")
```

```python
pd.merge(order,user,on = "user_id", how = "right")
```
