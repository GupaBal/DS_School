### 데이터 시각화 

```python
import pandas as pd 
```

### Load Dataset

```python
url = "https://bit.ly/dsa-04-health"  # 서울시에서 제공하는 공공데이터 임 

data = pd.read_csv(url)

print(data.shape)
data.head()
```

기본세팅하기.
import seaborn as sns    >>  데이터 시각화를 위한 import 및 설정 방법 

```python
import matplotlib
import seaborn as sns
from IPython.display import set_matplotlib_formats # 시각화를 뚜렸하게 하기위함 
#windows 
#matplotlib.rc('font', family = 'NanumGothic')
matplotlib.rc('font', family = 'Malgun Gothic')

# Retina 
set_matplotlib_formats('retina')

# Minus
matplotlib.rc('axes', unicode_minus = False)
```

```python
sns.countplot(data = data, x = "성별코드")     # 시각화 데이터가 불러짐  
```

### 사이즈 조절하기

###  distplot  (분포를 시각화 하는 것)

응용할 수 있는 것 알아보기

```python
# data["흡연상태"] 로 현재 흡연 상태를 알 수 있다. 
# 1이면 흡연 안 함, 2이면 흡연 했다가 그만 둠, 3이면 흡연 중 
data["흡연상태"].head(5)
```

```python
# 데이터 프레임 형식으로 1 조건 흡연 안 하는 사람들을 찾아 볼 수 있다. 
data[data["흡연상태"] == 1].head(5)
```

### 그래프로 흡연상태에 따른 허리둘레 비교해보기 

```python
one = data[data["흡연상태"] == 1]
two = data[data["흡연상태"] == 2]
three = data[data["흡연상태"] == 3]

sns.distplot(one["허리둘레"], hist = False, label = "흡연 안 함")
sns.distplot(two["허리둘레"], hist = False, label = "흡연 했지만 그만 둠")
sns.distplot(three["허리둘레"], hist = False, label = "흡연 하고 있음")
```

## countplot을 가져와서 응용하는 >>> barplot 
countplot = 데이터의 개수를 세지만 y축을 지정할 수 없다                              
barplot = y축을 지정할 수 있어진다. 조건은 숫자여야만 한다 (숫자의 평균을 구해줌) 

```python
plt.figure(figsize = (18,3))
sns.barplot(data = data, x = "연령대코드(5세단위)", y = "수축기혈압")
```

```python
# barplot는 estimator를 사용한다 

import numpy as np 


plt.figure(figsize = (18,3))
sns.barplot(data = data, x = "연령대코드(5세단위)", y = "수축기혈압", estimator = np.sum)
```

### 그래프의 X를 Y와 뒤집는 예시

```python
# X축 그래프
plt.figure(figsize = (18,3))
sns.barplot(data = data, x = "연령대코드(5세단위)", y="수축기혈압", hue = "성별코드")
```

```python
# Y축 그래프로 변경 
plt.figure(figsize = (4,9))
sns.barplot(data = data, y = "연령대코드(5세단위)", x ="수축기혈압", hue = "성별코드", orient = 'h')
```

### boxplot 가장 핫 한 시각화 중에 하나다 
시각화 되는 모양에 따라 여러가지 의미가 있으므로 사용할 때 한번 더 찾아보고 활용하도록~!!

```python
plt.figure(figsize = (18,3))
sns.boxplot(data = data, x = "연령대코드(5세단위)", y = "수축기혈압")
```

```python
plt.figure(figsize = (18,3))
sns.boxplot(data = data, x = "연령대코드(5세단위)", y = "수축기혈압", hue="성별코드")
```

### violinplot 핫 하고 boxplot과 비슷 하다
boxplot보다 데이터의 분포까지 보고싶을때 디테일 하게 보고싶을 때 사용한다.                 
일반적으로는 boxplot를 더 사용한다

```python
plt.figure(figsize = (18,3))
sns.violinplot(data = data, x = "연령대코드(5세단위)", y = "수축기혈압")
```

```python
plt.figure(figsize = (18,3))
sns.violinplot(data = data, x = "연령대코드(5세단위)", y = "수축기혈압", hue="성별코드")
```

```python
# 위에보다 좀 더 이쁘게 보고 싶을때 남성여성을 따로가 아니라 하나로 합쳐서 보여줌 
plt.figure(figsize = (18,3))
sns.violinplot(data = data, x = "연령대코드(5세단위)", y = "수축기혈압", hue="성별코드", split=True)
```

### pointplot 언제쓰냐? X축이 서로간에 연관성이 있을경우 (순서간에 )             
선을 그리면서 상관관계를 보여준다.

```python
plt.figure(figsize = (18,3))
sns.pointplot(data = data, x = "연령대코드(5세단위)", y = "수축기혈압", hue="성별코드")
```

```python
#선의 옵션을 줄 수도 있다 
plt.figure(figsize = (18,3))
sns.pointplot(data = data, x = "연령대코드(5세단위)", y = "수축기혈압", hue="성별코드",linestyles=[':','--'])
```

### scatterplot 요것도 요즘 많이 사용하는 것이다.  (점을 찍어서 데이터를 나타낸다)
순차적으로 사용하는 방법 

```python
# 초기 적용했을 때 
sns.scatterplot(data = data, x = "식전혈당(공복혈당)", y = "수축기혈압")
```

```python
# 150 아래만 살리고 이상은 지우기 
target = data[data["식전혈당(공복혈당)"] < 150]

sns.scatterplot(data = target, x = "식전혈당(공복혈당)", y = "수축기혈압")
```

```python
# 허리둘레를 추가하고, 원의 크기도 조절 할 수 있다 

plt.figure(figsize = (18,18))
target = data[data["식전혈당(공복혈당)"] < 150]

sns.scatterplot(data=target, x="식전혈당(공복혈당)", y="수축기혈압", hue="허리둘레", size="총콜레스테롤",sizes=(30, 550))
```
