```python
order["VIP"]= order["price"] >= 500   #이렇게 하면 price가 500이상인 조건 들로 VIP라는 컬럼이 형성됨 
order
```

```python
#만약에 두가지 이상의 조건으로 만든 컬럼이라면?
# price가 500 이상이면서 state가 confirm 조건인 것으로 VIP컬럼 생성 
order["VIP"] = (order["price"] >= 500) & (order["state"] == "confirm")
order
```

```python
#여러가지 조건을 추가하여 컬럼 생성 
order.loc[(order["price"] >= 500) & (order["state"] == "confirm"),"status"] = "VIP"
order.loc[(order["price"] < 500) | (order["state"] == "canceled"),"status"] = "None-VIP"
order
```

```python
# apply(?) 안에다가 함수같은걸 넣어서 for문 반복문처럼 돌린다. 
#def is_vip(price):
        # 여기 안에서는 판다스에서 실행할수 없는 구문을 파이썬 코드로 작서앟여 실행한다. 
        # 판다스로 구현하기 에매할 경우 이렇게 실행하기도 한다 
#    return price


def is_vip(price):
    if price >= 500:
        return "VIP"
    else: 
        return "Vone-VIp"

order["status"]= order["price"].apply(is_vip)
order
```

```python
#apply를 사용할때 여러개의 컬럼을 사용할 경우는 axis를 사용한다. 
order
```

```python
def is_vip(row):
    price = row["price"]
    state = row["state"]
    
    if price >= 500 and state == "confirm":
        return "VIP"
    else:
        return "None-VIP"
    
    return state
    
order["status"] = order.apply(is_vip, axis = "columns")
order
```

```python
# if ~ while 문 
coffee = 10
while True:
    money = int(input("돈을 넣어 주세요:"))
    if money == 300:
        print("커피를 줍니다.")
        coffee = coffee -1
    elif money > 300:
        print("거스름돈 %d를 주고 커피를 줍니다."% (money - 300))
        coffee = coffee - 1
    else:
        print("돈을 다시 돌려주고 커피를 주지 않습니다.")
        print("남은 커피의 양은 %d 개 입니다."% coffee)
    if coffee == 0:
        print("커피가 다 떨어졌습니다. 판매를 중지 합니다.")
        break
```
