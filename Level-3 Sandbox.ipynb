###  3-9 sql에서 부분적으로 컬럼을 가져올 때

```python
query = "SELECT id,date,price FROM 'orders'"  #디폴트 쿼리  orders쿼리에서 전체를 다 가져와라 

order = pd.read_sql(query, connection, index_col = "id")   #id를 인덱스의 디폴트로 지정하기
order.head()
```

```python
# 컬럼이 아니라 row를 가져올 때 
query = "SELECT * FROM 'orders' LIMIT 3"   # 일부분을 가져올 때  LIMIT 3 (수 많은 데이터를 가져올 수 없을 경우) 

order = pd.read_sql(query, connection, index_col = "id")   #id를 인덱스의 디폴트로 지정하기
order.head()
```

```python
# 시간 순으로 정렬되는 ORDER BY date                      
query = "SELECT * FROM 'orders' ORDER BY date"    

order = pd.read_sql(query, connection, index_col = "id")   #id를 인덱스의 디폴트로 지정하기
order.head()
```

sql은 단순 문법이 중요한게 아니라 여러가지 문법을 조화 했을때 능력이 나타난다       
여러가지 sql 문장을 사용하는 방법 
금액이 높은것 상위 3개를 추출해 내는 쿼리 

```python
query = """
    SELECT *
        FROM 'orders'
        ORDER BY price DESC
        LIMIT 3
"""

order = pd.read_sql(query, connection, index_col = "id")
order
```

### sql 문에서 가장 막강한  WHERE 문 = 색인 기능 및 검색기능 
ex) "SELECT * FROM 'orders' WHERE price >= 500"   조건문 가능 비교문 가능                      
ex) "SELECT * FROM 'orders' WHERE price <> 500" != 요거보다 <> 요걸 많이 사용한다 같은 뜻 인데 호환성 에서                   
ex) "SELECT * FROM 'orders' WHERE price BETWEEN 400 AND 900"  특정 범위안에서 찾고 싶을 때 사용하는 쿼리문   
ex) "SELECT * FROM 'orders' WHERE address IN ('Seoul'. 'Incheon')"

```python
query = "SELECT * FROM 'orders' WHERE price BETWEEN 400 AND 900 LIMIT 3 "

order = pd.read_sql(query, connection, index_col = "id")
order
```

```python
# 주소가 서울이거나 인천을 찾을 때 
order = pd.read_sql(query, connection, index_col = "id")
order
order = pd.read_sql(query, connection, index_col = "id")
order
```

"SELECT * FROM 'orders' WHERE price >= 500 AND state = 'confirmed'"          <<<< AND 조건       
"SELECT * FROM 'orders' WHERE price >= 500 OR state = 'confirmed'"          <<<< OR 조건
가격이 500이상 state 상태가 confirmed 

```python
query = """
    SELECT *
    FROM 'orders'
    WHERE price >= 500
    AND state = 'confirmed'
    ORDER BY price DESC
    LIMIT 3
"""

order = pd.read_sql(query, connection, index_col = "id")
order
```

### 여러 개의 데이터를 하나로 합치기   pandas의  merge 기능과 비슷 함 

```python
query = """
    SELECT *
        FROM 'orders'
    LEFT JOIN 'users' ON 'orders'.user_id = 'users'.id

"""
# 왼쪽이 orders 오른쪽이 users.id 
pd.read_sql(query, connection)
```

```python
#위에서와 같이 merge를 하게 될 경우 중복되는 데이터가 발생 한다 SELECT 에서 *을 했기 때문에 
#그래서 필요한 컬럼을 지정해서 불러오는게 좋다 

query = """
    SELECT 'orders'.id, 'orders'.date, 'orders'.price, 'users'.name, 'users'.gender
        FROM 'orders'
    LEFT JOIN 'users' ON 'orders'.user_id = 'users'.id

"""

pd.read_sql(query, connection)
```

```python
# INNER도 있다 이것도 교집합 
query = """
    SELECT 'orders'.id, 'orders'.date, 'orders'.price, 'users'.name, 'users'.gender
        FROM 'orders'
    INNER JOIN 'users' ON 'orders'.user_id = 'users'.id

"""

pd.read_sql(query, connection)
```
